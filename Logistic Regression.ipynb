```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
data=pd.read_csv("bank-full[1].csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data_1=pd.get_dummies(data,columns=['job','marital','education','contact','poutcome','month'])
data_1
```

```python
data_1.info()
```

```python
pd.set_option('display.max.columns',None)
data_1
```

```python
data_1=data_1.rename({'Target':'y'},axis=1)
data_1
```

```python
data_1['default']=np.where(data_1['default'].str.contains('yes'),1,0)
data_1['housing']=np.where(data_1['housing'].str.contains('yes'),1,0)
data_1['loan']=np.where(data_1['loan'].str.contains('yes'),1,0)
data_1['y']=np.where(data_1['y'].str.contains('yes'),1,0)
data_1
```

```python
data_1.info()
```

```python
x=pd.concat([data_1.iloc[:,0:10],data_1.iloc[:,11:]],axis=1)
y=data_1.iloc[:,10]
```

```python
classifier=LogisticRegression()
classifier.fit(x,y)
```

```python
y_=classifier.predict(x)
y_
```

```python
y_df=pd.DataFrame({'Actual_y':y,'y_predict':y_})
y_df
```

```python
confusion_matrix=confusion_matrix(y,y_)
confusion_matrix
```

```python
(39219+1047)/(39219+703+4242+1047)
```

```python
classifier.predict_proba(x)[:,1]
```

```python
fpr,tpr,thresholds=roc_curve(y,classifier.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='red')
auc=roc_auc_score(y,y_)

plt.plot(fpr,tpr,color='red',label='logistic model(area=%0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel("False positive rate")
plt.ylabel('True positive rate')
plt.show()
```

```python
print('auc accuracy:',auc)
```
